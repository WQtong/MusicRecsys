```python
listened_users = set(data_listen.msno)
```

```python
with open("./files/song_corpus.txt", "w") as f:
    for user in tqdm(list(listened_users)):
        user_hist = data_listen[data_listen.msno == user].reset_index(drop=True)
        contexts = []
        source_tab = None
        songs = user_hist.song_id
        tabs = user_hist.source_system_tab
        for song, tab in zip(songs, tabs):
            if source_tab == tab:
                try: 
                    contexts[-1].append(song)
                except IndexError as e:
                    print(e)
            else:
                contexts.append([song])
                source_tab = tab

        for context in contexts:
            for i in context:
                f.write(i + " ")
            f.write("\n")
```
